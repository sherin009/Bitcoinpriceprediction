```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('BitcoinPrice.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.drop(['Date'],1,inplace=True)
```

```python
df.head()
```

```python
#predicting 30 days in future
p_days=30
df['Prediction']=df[['Price']].shift(-p_days)
```

```python
df.head()
```

```python
df.tail()
```

```python
#seperate x and Y
x=np.array(df.drop(['Prediction'],1))
x = x[:len(df)-p_days]
```

```python
x.shape
```

```python
#creating y as np array
y=np.array(df['Prediction'])
y=y[: -p_days]
```

```python
y.shape
```

```python
y
```

```python
#split the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
```

```python
P_days_array = np.array(df.drop(['Prediction'], 1)[-p_days:])
```

```python
P_days_array
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
Rf=RandomForestRegressor(n_estimators=1000,random_state=1)
Rf.fit(x_train,y_train)
print('Random Forest Accuracy: {:2f}%'.format(Rf.score(x_test,y_test)*100))
```

```python
#prediction
Rf_predict=Rf.predict(x_test)
print(Rf_predict)
```

```python
print(y_test)
```

```python
#model prediction for 30 days
Rf_predict_30=Rf.predict(P_days_array)
print(Rf_predict)
```
